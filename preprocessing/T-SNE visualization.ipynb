```python
from sklearn.manifold import TSNE
```

```python
import pickle
import numpy as np

def get_similarity(a, b):
    dist = np.linalg.norm(a-b)
    return dist
    

f = open("embeddings.obj", 'rb')
embeddingList = pickle.load(f)
```

```python
key, attribute = zip(*[(line[0], line[1]) for line in embeddingList])
attribute = [l[0] for l in attribute]
attribute
```

```python
X_embedded = TSNE(n_components=3).fit_transform(attribute)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X_embedded)
```
